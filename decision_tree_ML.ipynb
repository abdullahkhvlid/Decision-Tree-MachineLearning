```python
import pandas as pd
df = pd.read_csv('salaries.csv')
df.head()
```

```python
input = df.drop('salary_more_then_100k', axis=1)
target = df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
input['nc'] = le_company.fit_transform(input['company'])
input['nj'] = le_company.fit_transform(input['job'])
input['nd'] = le_company.fit_transform(input['degree'])
```

```python
inputs = input.drop(['company', 'job', 'degree'], axis = 1)
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion="entropy", max_depth=10, random_state=42, min_samples_split=2)
model.fit(X_train, y_train)
```

```python
model.predict([[2,2,0]])
model.score(inputs, target)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
